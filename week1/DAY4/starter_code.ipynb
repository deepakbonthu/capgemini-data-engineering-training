```sql
%python
df1 = spark.read.csv("/Volumes/capgemini/default/task1/Capgemini Databricks VITB-2026 (1).csv", header=True, inferSchema=True)

df2 = spark.read.csv("/Volumes/capgemini/default/task1/Task 1 (Responses).csv", header=True, inferSchema=True)

df3 = spark.read.csv("/Volumes/capgemini/default/task1/Task 1 file 2.csv", header=True, inferSchema=True)
```

```sql
%python
df1.createOrReplaceTempView("master_raw")
df2.createOrReplaceTempView("task1_responses_raw")
df3.createOrReplaceTempView("task1_file2_raw")
```

```sql
SELECT * FROM master_raw;
```

```sql
CREATE OR REPLACE TEMP VIEW master_clean AS
SELECT *,
       LOWER(TRIM(`College Domain MAILID`)) AS college_email_clean,
       LOWER(TRIM(`Personal Mail ID`)) AS personal_email_clean
FROM master_raw;
```

```sql
CREATE OR REPLACE TEMP VIEW student_email_map AS
SELECT 
    `Superset ID` AS student_id,
    college_email_clean AS email
FROM master_clean

UNION

SELECT 
    `Superset ID` AS student_id,
    personal_email_clean AS email
FROM master_clean;
```

```sql
SELECT * FROM student_email_map LIMIT 10;
```

```sql
CREATE OR REPLACE TEMP VIEW task1_responses_clean AS
SELECT *,
       LOWER(TRIM(`Email address`)) AS email_clean
FROM task1_responses_raw;
```

```sql
SELECT * FROM task1_responses_clean LIMIT 5;
```

```sql
CREATE OR REPLACE TEMP VIEW task1_file2_clean AS
SELECT *,
       LOWER(TRIM(`Email address`)) AS email_clean
FROM task1_file2_raw;
```

```sql
select * from task1_file2_clean limit(5);
```

```sql
CREATE OR REPLACE TEMP VIEW student_email_map AS
SELECT 
    `Superset ID` AS student_id,
    college_email_clean AS email
FROM master_clean

UNION

SELECT 
    `Superset ID` AS student_id,
    personal_email_clean AS email
FROM master_clean;
```

```sql
select * from student_email_map limit(10);
```

```sql
CREATE OR REPLACE TEMP VIEW responses_mapped AS
SELECT 
    r.*,
    m.student_id
FROM task1_responses_clean r
LEFT JOIN student_email_map m
ON r.email_clean = m.email;
```

```sql
select * from responses_mapped limit(10);
```

```sql
CREATE OR REPLACE TEMP VIEW file2_mapped AS
SELECT 
    f.*,
    m.student_id
FROM task1_file2_clean f
LEFT JOIN student_email_map m
ON f.email_clean = m.email;
```

```sql
CREATE OR REPLACE TEMP VIEW all_submissions AS
SELECT 
    student_id,
    email_clean,
    'responses' AS source
FROM responses_mapped

UNION ALL

SELECT 
    student_id,
    email_clean,
    'file2' AS source
FROM file2_mapped;
```

```sql
SELECT * FROM all_submissions LIMIT 10;
```

```sql
CREATE OR REPLACE TEMP VIEW submissions_with_rank AS
SELECT *,
       ROW_NUMBER() OVER (PARTITION BY student_id ORDER BY student_id) AS rn
FROM all_submissions;
```

```sql
SELECT * FROM submissions_with_rank LIMIT 10;
```

```sql
CREATE OR REPLACE TEMP VIEW final_student_status AS
SELECT 
    m.`Superset ID` AS student_id,
    CASE 
        WHEN s.student_id IS NULL THEN 'NOT_SUBMITTED'
        WHEN s.rn = 1 THEN 'VALID_SUBMISSION'
        ELSE 'DUPLICATE'
    END AS status
FROM master_clean m
LEFT JOIN submissions_with_rank s
ON m.`Superset ID` = s.student_id;
```

```sql
SELECT * FROM final_student_status;
```

```sql
SELECT status, COUNT(*) 
FROM final_student_status
GROUP BY status;
```
